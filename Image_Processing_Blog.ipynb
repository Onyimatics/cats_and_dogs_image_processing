# **Cat vs Dog Image Classification**

This project focuses on developing a machine learning model to classify images as either cats or dogs. The dataset consists of 10,000 labeled training images and 1,000 test images requiring classification.


## Data Preprocessing

Several models were experimmeted utilizing different techniques and processes. These processes are gray scaling, smoothing, image vectorization, image resizing, data augmentation. The first model experimented was SVM which achieved 66.96% using gray scaling, image resizing and feature extractio with hog. Logistic Regression achieved 68.6% on the same preproccessing techniques and thereafter Random Forest Classifier achieved 70.1%.

As we can see, the above accuracies are quite low so I decided to experiment with Convolutional Neural Network. For the neural network, six different image preprocessing techniques were explored to determine which would yield the best classification results:

1. **Original RGB**: Standard RGB normalization (dividing pixel values by 255)
2. **Grayscale**: Converting images to grayscale to reduce complexity
3. **Gaussian Blur**: Applying blur to grayscale images to reduce noise
4. **Canny Edge Detection**: Extracting edge information to focus on shapes
5. **HOG Features**: Using Histogram of Oriented Gradients to capture structural information
6. **Combined Method**: Merging grayscale and the above edge detection techniques

All images were resized to 150x150 pixels for consistency before preprocessing.

## Data Augmentation

To enhance model robustness and combat overfitting, data augmentation was applied to all preprocessing methods: Rotation (20 degrees), Width/height shifts (20%), Shear transformation (20%), Zoom variation (20%), Horizontal flipping and Brightness adjustments (for RGB images).

## Model Architecture

As seen below model architectural design, the model uses a progressive doubling of filters which is a common pattern in CNN architectures to increase feature extraction capacity as the spatial dimensions decrease through pooling.

## Results and Evaluation

After training the model on each preprocessing technique with data augmentation, the Grayscale preprocessing method achieved the highest validation accuracy of 80.80% successfully classifying cats and dogs with good confidence.
